### 4. best.pt 이어서 재학습

```python
from ultralytics import YOLO
from huggingface_hub import hf_hub_download

# 기존 best.pt 로드
model = YOLO(hf_hub_download("huggingleg12/recycle-yolo11", "weights/best.pt"))

model.train(
    data="data_v2.yaml",   # 새로운 어플리케이션 데이터
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=4,
    cls=2.0,
    hsv_v=0.6,
    hsv_s=0.6,
    degrees=20,
    perspective=0.001,
    save=True,
    save_period=10,
    project="/content/drive/MyDrive/yolo_recycle/runs",
    name="recycle_yolo11_v2",
)
```

### 5. 성능 평가

```python
model = YOLO("/content/drive/MyDrive/yolo_recycle/runs/recycle_yolo11_v2/weights/best.pt")
results = model.val(data="data_v2.yaml", device=0, split="test")

print(f"\n전체 mAP@50:    {results.box.map50:.4f}")
print(f"전체 mAP@50-95: {results.box.map:.4f}")

# 클래스별 mAP 출력
print("\n클래스별 mAP@50:")
names = ["종이","종이팩","종이컵","캔류","재사용유리","색깔유리","페트","플라스틱","비닐","스티로폼","건전지"]
for i, (name, ap) in enumerate(zip(names, results.box.ap50)):
    print(f"  {name}: {ap:.4f}")
```

### 6. 허깅페이스 업로드

```python
from huggingface_hub import HfApi
api = HfApi(token="토큰입력")

api.upload_file(
    path_or_fileobj="/content/drive/MyDrive/yolo_recycle/runs/recycle_yolo11_v2/weights/best.pt",
    path_in_repo="weights/best.pt",
    repo_id="huggingleg12/recycle-yolo11",
    repo_type="model",
    token="토큰입력"
)
print("업로드 완료!")
```

```python
from PIL import Image
import numpy as np
from google.colab import files

# 사진 업로드
uploaded = files.upload()
model = YOLO("/content/drive/MyDrive/yolo_recycle/runs/recycle_yolo11_v2/weights/best.pt")
for filename in uploaded.keys():
    img = Image.open(filename).convert("RGB")
    print(f"\n=== {filename} ===")
    print(f"원본 크기: {img.size}")

    # 640×640으로 리사이즈
    img_resized = img.resize((640, 640))
    results = model(np.array(img_resized), conf=0.1)

    print(f"탐지 수: {len(results[0].boxes)}")
    for box in results[0].boxes:
        label = model.names[int(box.cls)]
        conf = float(box.conf)
        print(f"클래스: {label}, 신뢰도: {conf:.3f}")

    results[0].show()
```
