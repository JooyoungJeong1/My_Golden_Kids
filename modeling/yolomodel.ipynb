```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11m.pt")
baseline = model.val(data="data.yaml", device=0, split="test")
print(f"베이스라인 mAP@50: {baseline.box.map50:.4f}")
print(f"베이스라인 mAP@50-95: {baseline.box.map:.4f}")
```

```python
model = YOLO("yolo11m.pt")
model.train(
    data="data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,
    workers=4,
    cls=2.0,
    save=True,
    save_period=10,
    project="/content/drive/MyDrive/yolo_recycle/runs",
    name="recycle_yolo11",
)
```

```python
model = YOLO("/content/drive/MyDrive/yolo_recycle/runs/recycle_yolo11/weights/best.pt")
results = model.val(data="data.yaml", device=0, split="test")
print(f"최종 mAP@50:    {results.box.map50:.4f}")
print(f"최종 mAP@50-95: {results.box.map:.4f}")
```

```python
!pip install huggingface_hub

from huggingface_hub import HfApi
from google.colab import userdata

token = userdata.get('HF_TOKEN_YOLO')
api = HfApi(token=token)

api.create_repo("recyclesight-yolo11", repo_type="model", exist_ok=True)

api.upload_file(
    path_or_fileobj="/content/drive/MyDrive/yolo_recycle/runs/recycle_yolo11/weights/best.pt",
    path_in_repo="weights/best.pt",
    repo_id="huggingleg12/recycle-yolo11",
    repo_type="model",
)
print("업로드 완료!")
```

```python
from ultralytics import YOLO
from huggingface_hub import hf_hub_download

model = YOLO(hf_hub_download("huggingleg12/recycle-yolo11", "weights/best.pt"))
print(model.names)
```
